```python
#fun for scraping the data from url
def scrape_title_and_text_from_url(url):

    browser.get(url)
    titles = []
    text_bodies = []

    # Scraping titles
    print('Scraping titles and text bodies from:', url)
    title_elements = browser.find_elements(By.XPATH, "//*[@class='entry-title']")
    for title in title_elements:
        titles.append(title.text)

    # Scraping articles text body
    body_elements = browser.find_elements(By.XPATH, "//*[@class='td-post-content tagdiv-type']")
    for body in body_elements:
        text_bodies.append(body.text)

    #browser.quit()  # Close the browser session
    return titles, text_bodies
```

```python
def add_stopwords(file_paths):
    nltk_stopwords = stopwords.words('english')  # Initialize NLTK stopwords list

    for file_path in file_paths:
        with open(file_path, 'r') as file:
            stopwords_list = file.readlines()

        # Convert stopwords to lowercase and remove leading/trailing whitespace
        stopwords_list = [word.strip().lower() for word in stopwords_list]

        # Extend the NLTK stopwords list with the custom stopwords

        nltk_stopwords.extend(stopwords_list)

    return nltk_stopwords

    print("Custom stopwords added:", nltk_stopwords)
```

```python
import nltk

#tokenize each paragr into sentences
def tokenize_into_sentences(paragraphs):
    # Tokenize each paragraph into sentences
    sentences = nltk.sent_tokenize(paragraphs)
    return sentences

#check
paragraphs = "I Rising IT Cities and Their Impact on the Economy, Environment, Infrastructure, and City Life in Future. Throughout history, from the industrial revolution in the 18th century through the development of the Internet, technology has been the primary driver of societal change."
sentences = tokenize_into_sentences(paragraphs)
print(sentences)
```

```python
# Function to preprocess sentences
def preprocess_sentences(sentences, custom_stopwords):
    stop_words = set(nltk.corpus.stopwords.words('english'))
    stop_words.update(custom_stopwords)

    preprocessed_sentences = []
    for sentence in sentences:
        # Remove characters other than letters
        sentence = re.sub(r'[^a-zA-Z]', ' ', sentence)
        # Convert the sentence to lowercase
        sentence = sentence.lower()
        # Tokenize the sentence into words
        words = nltk.word_tokenize(sentence)
        # Remove stopwords
        filtered_words = [word for word in words if word not in stop_words]
        # Join the words back into a sentence
        preprocessed_sentence = ' '.join(filtered_words)
        preprocessed_sentences.append(preprocessed_sentence)
    return preprocessed_sentences
```

```python
#$#$Function to read files, tokenize into sentences, and preprocess sentences
def preprocessing(input_dir, df, custom_stopwords):
    preprocessed_data = []
    for index, row in df.iterrows():
        url_id = row['URL_ID']
        url = row['URL']

        file_path = os.path.join(input_dir, f"{url_id}.txt")

        with open(file_path, 'r', encoding='utf-8') as file:
            paragraphs = file.read()

        # Tokenize paragraphs into sent
        sentences = tokenize_into_sentences(paragraphs)

        # Preprocess sentences
        preprocessed_sentences = preprocess_sentences(sentences, custom_stopwords)

        ##Append preprocessed data to the list
        preprocessed_data.append(preprocessed_sentences)

    # Add preprocessed data as a new column in df
    df['preprocessed_data'] = preprocessed_data

    return df
```

## 1.2 Creating a dictionary of Positive and Negative words

Function to read positive and negative words from text files

```python
def read_positive_negative_words(positive_file_path, negative_file_path):
    with open(positive_file_path, 'r', encoding='utf-8') as file:
        positive_words = set(file.read().splitlines())

    with open(negative_file_path, 'r', encoding='utf-8') as file:
        negative_words = set(file.read().splitlines())

    return positive_words, negative_words
```

Function to create dictionary of positive and negative words for each row of preprocessed data

```python
def create_word_sentiment_dictionary(preprocessed_data, positive_words, negative_words):
    word_sentiment_dict_list = []
    for sentences in preprocessed_data:
        word_sentiment_dict = {'positive_dict': {}, 'negative_dict': {}, 'neutral_dict': {}}
        for sentence in sentences:
            words = sentence.split()
            for word in words:
                if word in positive_words:
                    word_sentiment_dict['positive_dict'][word] = word_sentiment_dict['positive_dict'].get(word, 0) + 1
                elif word in negative_words:
                    word_sentiment_dict['negative_dict'][word] = word_sentiment_dict['negative_dict'].get(word, 0) + 1
                else:
                    word_sentiment_dict['neutral_dict'][word] = word_sentiment_dict['neutral_dict'].get(word, 0) + 1
        word_sentiment_dict_list.append(word_sentiment_dict)
    return word_sentiment_dict_list
```

## Polarity Score
    - This function takes the positive score and negative score as input and returns the calculated polarity score

```python
def polarity_score(positive_score, negative_score):
    denominator = (positive_score + negative_score + 0.000001)
    polarity_score = (positive_score - negative_score) / denominator
    return polarity_score
```

## Percentage of Complex words
    - complex words are considered to be words with more than a certain number of syllables
    - counting syllables in a word

calculating percentage of complex words

```python
def percentage_complex_words(words):
    num_complex_words = 0
    num_words = len(words)

    # Define a function to count syllables in a word
    def count_syllables(word):
        return max(1, len([char for char in word if char.lower() in 'aeiou']))

    # Count the number of complex words
    for word in words:
        if count_syllables(word) > 2:  # Adjust threshold for complexity as needed
            num_complex_words += 1

    # Calculate the percentage of complex words
    if num_words != 0:
        percentage_complex_words = (num_complex_words / num_words) * 100
    else:
        percentage_complex_words = 0

    return percentage_complex_words
```

```python
def syllable_count(word):
    if word.lower() in d:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]])
    else:
        ###Estimate syllables basesd on vowels
        return max(1, sum([1 for char in word if char.lower() in 'aeiou']))
```

```python
####count complex words in a text
def count_complex_words(text):
    words = nltk.word_tokenize(text)
    complex_words_count = sum(1 for word in words if syllable_count(word) > 2)
    return complex_words_count
```

# 6. Syllable Count Per Word

To count the number of syllables in each word of the text while handling exceptions like words ending with "es" or "ed"

```python
def count_syllables(word):
    vowels = 'aeiouy'
    # Remove any non-alphabetic characters from the word
    word = ''.join(filter(str.isalpha, word))
    # Check for exceptions
    exceptions = ['es', 'ed']
    for exception in exceptions:
        if word.endswith(exception):
            return max(1, len([char for char in word if char in vowels]))
    # Count the number of vowels
    syllables = len([char for char in word if char in vowels])
    return max(1, syllables)
```

```python
# Function to count total number of syllables in a text
def count_total_syllables(text):
    words = nltk.word_tokenize(text)
    total_syllables_count = sum(count_syllables(word) for word in words)
    return total_syllables_count
```

```python
df['syllables_per_word'] = df['preprocessed_data'].apply(lambda x: count_total_syllables(' '.join(x)))
```

# Personal Pronouns

To calculate the count of personal pronouns mentioned in the text while excluding the country name "US," you can use regular expressions.

```python
def count_personal_pronouns(text):
    # Define a regex pattern to match personal pronouns
    pattern = r'\b(?:I|we|my|ours|us)\b'
    # Compile the regex pattern
    regex = re.compile(pattern, flags=re.IGNORECASE)
    # Find all matches of the pattern in the text
    matches = regex.findall(text)
    # Return the count of matches
    return len(matches)
```

# Average Word Length

## To calculate the average word length in a text, you can follow these steps:

- Split the text into individual words.
- Calculate the total number of characters in all words.
- Count the total number of words.
- Divide the total number of characters by the total number of words to get the average word length.

```python
def average_word_length(text):
    words = text.split()
    # Split the text into words
    total_characters = sum(len(word) for word in words)

    # Calculate the total number of characters
    total_words = len(words)

    # Calculate the total number of words
    if total_words > 0:
        average_length = total_characters / total_words

        # Calculate the average word length
    else:
        average_length = 0
    return average_length
```
